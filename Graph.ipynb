```python
import numpy as np
import os 
os.environ['KMP_DUPLICATE_LIB_OK']='True'

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
episodes = np.arange(1, 2001)
DQN = np.load('All_Average_Rewards/DQN_avg_rewards.npy', allow_pickle=True)
DDQN = np.load('All_Average_Rewards/DDQN_avg_rewards.npy', allow_pickle=True)
DDPG = np.load('All_Average_Rewards/DDPG_avg_scores.npy', allow_pickle=True)
TD3 = np.load('All_Average_Rewards/TD3_avg_rewards.npy', allow_pickle=True)
```

```python
fig, ax = plt.subplots(figsize=(12, 6))

ax.plot(DQN, label='DQN Performance')
ax.plot(DDQN, label='DDQN Performance')

ax.set_xlabel('Episodes')
ax.set_ylabel('Reward')
ax.set_title('Rewards and Moving Average Rewards per Episode')
ax.legend()
ax.grid(True)

plt.show()
plt.savefig("./Lunar-Landerv1.0-DiscreteSpace.jpg")
```

```python
fig, ax = plt.subplots(figsize=(12, 6))

ax.plot(DDPG, label='DDPG Performance')
ax.plot(TD3, label='TD3 Performance')

ax.set_xlabel('Episodes')
ax.set_ylabel('Reward')
ax.set_title('Rewards and Moving Average Rewards per Episode')
ax.legend()
ax.grid(True)

plt.show()
```
